```python
def solve(s):
    toDelete = s[0]
    others = ""
    for c in s:
        if c != toDelete:
            others += c
    s = others
#prefix of length 1 surely doesn't contain different letters
    prefix = s[0]
    for i in range(1, len(s)):
#Schlumberger-Private#each letter should be the same as the first
        if s[i] == prefix[0]:
            prefix += s[i]
        else:
            break
    return prefix
solve("ttttt")
```
